```python
import matplotlib.pyplot as plt

# Data for the bar chart
features = ['Basic', 'GIST', 'LBP', 'GLCM', 'Synergy']
times = [8.7, 27.4, 8.1, 8.8, 34.4]

# Creating the bar chart
plt.figure(figsize=(10, 6))  # Adjust the size to your preference
bars = plt.bar(features, times, color=['blue', 'orange', 'green', 'red', 'purple'])

# Adding the text on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f} s', ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Type of feature')
plt.ylabel('Time (s)')
plt.title('Comparison of Feature Computation Times')

# Display the bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
features = ['GIST', 'GLCM', 'LBP', 'Basic', 'Synergy']
times = [27.40, 8.80, 8.10, 8.70, 34.4]

# Creating the bar chart
plt.figure(figsize=(10, 3))  # Adjust the size to your preference
bars = plt.bar(features, times, color=['blue', 'orange', 'green', 'red', 'purple'])

# Adding the text on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f} s', ha='center', va='bottom', fontsize=12)

# Adding labels and title with larger font size
plt.xlabel('Type of feature', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.ylim(0, 40)
# plt.title('Comparison of Feature Computation Times', fontsize=16)

# Increasing the font size for the x-axis feature names
plt.xticks(fontsize=13.5)

# Display the bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data for the bar chart
features = ['GIST', 'GLCM', 'LBP', 'Basic', 'Synergy']
times = [27.40, 8.80, 8.10, 8.70, 34.4]

# Using Seaborn style without grid lines
plt.figure(figsize=(10, 3))
sns.set(style="darkgrid")


# Creating the bar chart with Seaborn
bars = sns.barplot(x=features, y=times, palette="muted")

# Adding the text on top of each bar
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f} s', ha='center', va='bottom', fontsize=12)

# Adding labels and title with larger font size
plt.xlabel('Type of feature', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.ylim(0, 40)

# Increasing the font size for the x-axis feature names
plt.xticks(fontsize=13.5)

# Remove top and right spines
sns.despine()

# Display the bar chart
plt.show()
```
